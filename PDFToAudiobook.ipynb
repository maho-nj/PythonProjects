## Create an Audiobook from a PDF file with Python

```
#Let's start by installing PyPDF2 and pyttsx3. pYpdf2 enables the manipulation of PDF in memory. Such manipulation include:
#encryption and decryption PDF files,
#extraction of relevant information from the document, 
#merging documents per page
#dividing document by page
# or merging multiple pages into one page, etc.
#While the pyttsx3 has the ability of converting text to speech offline.

!pip install PyPDF2
!pip install pyttsx3
```

```
#The following lines of code divide the PDF file page by page, then read the text on each page,
#then send the text to the next step in the process to create an audiobook with Python.
import PyPDF2 

pdfReader = PyPDF2.PdfFileReader(open('/content/drive/MyDrive/StudyDocuments/DataScienceKeyTermsAndConcepts/Python Projects/Biogas_2008.pdf', 'rb'))
```

```
# Now, the text that we read from the PDF file is fed as an input to the text-to-speech engine
!sudo apt install libespeak1
import pyttsx3
speaker = pyttsx3.init()
```

```
## Then, let's loop the process for each page of the PDF file and stop the pyttsx3 speaker engine last:
speaker.setProperty('rate',70)
voices=speaker.getProperty('voices')

for page_num in range(pdfReader.numPages):
    text =  pdfReader.getPage(page_num).extractText()
    speaker.say(text)
    speaker.runAndWait()
speaker.stop()
```

```
## At last, let's save the audio as an mp3.file
speaker.save_to_file(text, 'audio.mp3')
speaker.runAndWait()
```
