## BMI Calculator 

This script calculates and comments your body mass index. The BMI is given by the following formula: BMI = m/h^2 

Where m is the mass in kilograms and h is the height in meter. 

```
while True:
  try:
    H=input("Enter your height in feet, inches: ")
    Weight=float(input("Enter your Weight in lbs: "))
    height=H.split(',')
    feet = int (height[0])
    inches= int (height[1])
    Height = (12*feet + inches)*0.0254
    Weight = Weight*0.453592
    BMI=int(Weight/(Height*Height))
    WTL = int(((BMI-25)*(Height*Height))/0.453592)
    break 
  except ValueError:
    print ('Oops! An incorrect value was entered. Please make sure your height and weight are entered correctly. Try again ...')
  except IndexError:
    print ("One on the values you've entered is out of range, please try again ...")


print("Your Body Mass Index is: ",BMI)
if(BMI>0):
  if(BMI<=16):
    print("You are severely underweight, you need to gain this weight in lbs to become healthy: ", -1*WTL)
  elif(BMI<=18.5):
    print("You are underweight, you need to gain this weight in lbs to become healthy: ", -1*WTL)
  elif(BMI<=25):
    print("You are healthy. Keep it up!")
  elif(BMI<=30):
    print("You are overweight. You need to lose this weight in lbs to become healthy: ", WTL)
  else: print("you are severely overweight. You need to lose this weight in lbs to become healthy:", WTL)
else:("enter valid details")
```
