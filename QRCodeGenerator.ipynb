#QR Code Generator

QR codes are used to code and decode data into into a machine learning form. QR codes contain grid black squares on a white background,which can be read by any imaging device, which can extract the required data from the patterns of the QR code.

```
!pip install pyqrcode
```

```
import pyqrcode
from pyqrcode import QRCode 
import os 


website = str(input("What's the link to the homepage of your website?"))
name=str(input("How do you want to name your file?"))
suffix =".svg"
url = pyqrcode.create(website)
filename = os.path.join(name+suffix)
url.svg(filename, scale =8)
```
